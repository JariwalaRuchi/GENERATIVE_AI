#**Loading Model**

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2-large")
model = GPT2LMHeadModel.from_pretrained("gpt2-large", pad_token_id=tokenizer.eos_token_id)
```

#**Preprocessing The Data**

```python
text = "what is the Natural language processing?"
encoded_input = tokenizer.encode(text, return_tensors='pt')
```

```python
encoded_input
```

```python
tokenizer.decode(encoded_input[0][0])
```

#**Text Generation**

```python
output = model.generate(encoded_input, max_length=500, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
```

```python
output
```

#**Decoding The encoded text generated by model**

```python
tokenizer.decode(output[0], skip_special_tokens=True)
```
